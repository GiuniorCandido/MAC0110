Introdução
Três números inteiros positivos a, b e c, a < b < c, formam um trio Pitagoreano se a2 + b2 = c2 . Por exemplo, os números 3, 4, e 5 formam um trio Pitagoreano pois 32 + 42 = 52.

Alguns números inteiros positivos podem ser escritos como a soma de um trio Pitagoreano. Por exemplo, 12 é um desses números pois 3 + 4 + 5 = 12.

Escreva um programa em Python que lê um número inteiro positivo n e verifica se ele é soma de um trio Pitagoreano. Em caso afirmativo, o seu programa deve imprimir os valores do trio e, em caso contrário, deve imprimir que o número não é soma de um trio Pitagoreano.

Exemplos
A seguir estão 4 exemplos de execução do programa. A saída do seu programa deve ser igual às mostradas nos exemplos. Os valores em vermelho foram digitados pelo usuário e em verde é a saída do programa.

10 não é soma de trio Pitagoreano

Digite o valor de n: 12

12 é soma do trio (3, 4, 5)

Digite o valor de n: 24

24 é soma do trio (6, 8, 10)

Digite o valor de n: 392

392 é soma do trio (49, 168, 175)
